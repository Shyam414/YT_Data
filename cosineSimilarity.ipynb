```python
import pandas as pd
import demoji
import re
from collections import Counter
import math
```

```python
df=pd.read_csv('Data\\commentsData.csv')
df
```

```python
def remove_hashtags(title):
    return ' '.join(word for word in title.split() if not word.startswith('#'))

# Apply the function to the 'title' column
df['title'] = df['title'].apply(remove_hashtags)
df['title']
```

```python
def remove_emojis(text):
    return demoji.replace(text, '')

df['comments'] = df['comments'].apply(remove_emojis)
df
```

```python
def clean_text(text):
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    text = text.lower()
    return text
df['comments'] = df['comments'].apply(lambda x: clean_text(x))

df
```

```python
def preprocess_data(df):
    df = df.drop_duplicates(subset=['comments'])
    
    spam_keywords = ['spam', 'advertisement', 'fake']
    df = df[~df['comments'].str.contains('|'.join(spam_keywords), case=False)]
        
    return df
df = preprocess_data(df)
df
```

```python
comments_list = df['comments'].tolist()
comments_list
```

```python
def cosine_similarity(user_question, comments_list):
    user_tokens = user_question.lower().split()
    similarity_scores = []

    for comment in comments_list:
        comment_tokens = comment.lower().split()

        user_counter = Counter(user_tokens)
        comments_counter = Counter(comment_tokens)   

        dot_product = sum(user_counter[token] * comments_counter[token] for token in user_counter.keys())

        user_magnitude = math.sqrt(sum(user_counter[token] ** 2 for token in user_counter))
        comments_magnitude = math.sqrt(sum(comments_counter[token] ** 2 for token in comments_counter))

        similarity = dot_product / (user_magnitude * comments_magnitude) if user_magnitude * comments_magnitude != 0 else 0
        similarity_scores.append(similarity)

    return similarity_scores
```

```python
def return_response(query, corpus):
    similarities = []
    for doc in corpus:
        similarity = cosine_similarity(query, doc)
        similarities.append(similarity)
    return comments_list[similarities.index(max(similarities))]
```

```python
user="which video is most liked one"

relevant_document=return_response(user,comments_list)
relevant_document
```
